```python
import pyvo, pyvo.registry
```

```python
from astropy import coordinates, units as u
```

```python
src = coordinates.SkyCoord('05:24:13.900 -03:37:00.300', frame='fk5', unit=(u.h, u.deg))
```

```python
for svc in pyvo.registry.search(servicetype="image", waveband="optical"):
  try:
    images = svc.search(pos=src, size=5*u.arcmin)
    print(images)
  except Exception:
    import traceback; traceback.print_exc()
```
